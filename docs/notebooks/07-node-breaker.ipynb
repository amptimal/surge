# Node-Breaker Topology

Companion notebook for Tutorial 07. This notebook needs user-supplied CGMES or XIIDM files that retain node-breaker topology.


```python
import os
import surge

eq = os.environ.get("SURGE_CGMES_EQ")
tp = os.environ.get("SURGE_CGMES_TP")
ssh = os.environ.get("SURGE_CGMES_SSH")

if not (eq and tp and ssh):
    print("Set SURGE_CGMES_EQ, SURGE_CGMES_TP, and SURGE_CGMES_SSH to run this notebook.")
else:
    net = surge.load(eq, tp, ssh)
    topology = net.topology
    print("status:", topology.status if topology is not None else "missing")
    if topology is not None and topology.switches:
        first = topology.switches[0]
        topology.set_switch_state(first.id, is_open=not first.is_open)
        rebuilt = topology.rebuild()
        print("rebuilt buses:", rebuilt.n_buses)
        print("mapping bus count:", rebuilt.topology.current_mapping().n_buses if rebuilt.topology is not None else "n/a")
```
