# Contingency Analysis

Companion notebook for Tutorial 02. Compare screening modes and reusable study objects.


```python
import surge

net = surge.case118()
lodf = surge.analyze_n1_branch(net, surge.ContingencyOptions(screening="lodf", thermal_threshold_pct=100.0))
full = surge.analyze_n1_branch(net, surge.ContingencyOptions(screening="off", thermal_threshold_pct=100.0))
n2 = surge.analyze_n2_branch(net, surge.ContingencyOptions(screening="lodf", top_k=100, thermal_threshold_pct=100.0))
study = surge.contingency.n1_branch_study(net, surge.ContingencyOptions(screening="lodf", thermal_threshold_pct=100.0))
repeat = study.analyze()

print("N-1 LODF:", lodf.n_contingencies, lodf.n_with_violations, lodf.solve_time_secs)
print("N-1 full:", full.n_contingencies, full.n_with_violations, full.solve_time_secs)
print("N-2:", n2.n_contingencies, n2.n_with_violations, n2.solve_time_secs)
print("Study rerun:", repeat.n_contingencies, repeat.n_with_violations)
```
