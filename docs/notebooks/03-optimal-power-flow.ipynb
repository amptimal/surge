# Optimal Power Flow

Companion notebook for Tutorial 03. Compare DC-OPF, AC-OPF, and SCOPF on the same network.


```python
import surge

net = surge.case118()
dc = surge.solve_dc_opf(net, options=surge.DcOpfOptions(enforce_thermal_limits=True), runtime=surge.DcOpfRuntime(lp_solver="highs"))
ac = surge.solve_ac_opf(net, options=surge.AcOpfOptions(enforce_thermal_limits=True), runtime=surge.AcOpfRuntime(nlp_solver="ipopt", exact_hessian=True, print_level=0))
scopf = surge.solve_scopf(net, options=surge.ScopfOptions(formulation=surge.ScopfFormulation.DC, mode=surge.ScopfMode.PREVENTIVE, contingency_rating=surge.ThermalRating.RATE_A), runtime=surge.ScopfRuntime(lp_solver="highs", max_iterations=10))

print("DC cost:", dc.total_cost)
print("AC cost:", ac.total_cost)
print("SCOPF cost:", scopf.total_cost)
print("First 5 DC LMPs:", dc.lmp[:5])
print("First 5 AC LMPs:", ac.lmp[:5])
```
