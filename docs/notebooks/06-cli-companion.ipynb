# CLI Companion

Companion notebook for Tutorial 06. Run `surge-solve` from Python when you want to inspect CLI behavior interactively.


```python
from pathlib import Path
import subprocess

from workflow_common import find_case, find_cli

cli = find_cli()
case_path = find_case("case118.surge.json.zst")
print("cli:", cli)
print("case:", case_path)

if cli is None or case_path is None:
    print("Build the CLI and ensure a sample case file is available before running this notebook.")
else:
    result = subprocess.run([str(cli), str(case_path), "--method", "acpf", "--quiet"], capture_output=True, text=True)
    print("returncode:", result.returncode)
    print("stdout:")
    print(result.stdout[:400])
    print("stderr:")
    print(result.stderr[:400])
```
