# Basic Power Flow

Companion notebook for Tutorial 01. Compare the main AC solve, the DC approximation, and FDPF on the same case.


```python
import surge

net = surge.case118()
ac = surge.solve_ac_pf(net, surge.AcPfOptions(tolerance=1e-8, flat_start=False, dc_warm_start=True))
ac_flat = surge.solve_ac_pf(net, surge.AcPfOptions(tolerance=1e-8, flat_start=True, dc_warm_start=False))
fdpf = surge.powerflow.solve_fdpf(net, surge.powerflow.FdpfOptions(tolerance=1e-6, flat_start=True))
dc = surge.solve_dc_pf(net)

print("AC:", ac.converged, ac.iterations, ac.max_mismatch)
print("AC flat:", ac_flat.converged, ac_flat.iterations, ac_flat.max_mismatch)
print("FDPF:", fdpf.converged, fdpf.iterations, fdpf.max_mismatch)
print("DC solve time:", dc.solve_time_secs)
print("First 5 AC angles (deg):", ac.va_deg[:5])
print("First 5 DC angles (rad):", dc.va_rad[:5])
```
